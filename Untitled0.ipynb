```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
clients = pd.read_csv("/content/clients.csv")
properties = pd.read_csv("/content/properties(2).csv")

print("Clients Shape:", clients.shape)
print("Properties Shape:", properties.shape)

clients.head()
```

```python
# Merge clients and properties

df = pd.merge(
    clients,
    properties,
    left_on="client_id",
    right_on="client_ref",
    how="inner"
)

print("Merged Dataset Shape:", df.shape)

df.head()
```

```python
# Remove duplicate rows
df = df.drop_duplicates()

print("Dataset Shape:", df.shape)
```

```python
# Convert mixed date formats
df['date_of_birth'] = pd.to_datetime(
    df['date_of_birth'],
    format='mixed',
    errors='coerce'
)

# Create Age column
today = pd.Timestamp.today()
df['Age'] = ((today - df['date_of_birth']).dt.days / 365.25).astype(int)

# Check result
df[['date_of_birth', 'Age']].head()
```

```python
df = df.drop(columns=[
    'first_name',
    'last_name',
    'date_of_birth'
])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_columns = df.select_dtypes(include='object').columns

for col in categorical_columns:
    df[col] = le.fit_transform(df[col].astype(str))

df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_data = scaler.fit_transform(df)

print("Scaled Data Shape:", scaled_data.shape)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)

df['Cluster'] = kmeans.fit_predict(scaled_data)

df.head()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(scaled_data, df['Cluster'])

print("Silhouette Score:", score)
```
